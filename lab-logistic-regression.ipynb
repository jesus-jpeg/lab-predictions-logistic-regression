```python
# Basic libraries
import os # file management 
import pandas as pd # data manipulatioN
import numpy as np # numerical operations
import matplotlib.pyplot as plt # 2D visualizations
import seaborn as sns # advanced statistical visualization
import warnings # warning messages management

# SQL connection
from sqlalchemy import create_engine # let's connect to SQL
from getpass import getpass
password = getpass("Please, kindly insert your password:")

# Machine Learning
from sklearn.preprocessing import StandardScaler # feature scaling
from sklearn.model_selection import train_test_split # splitting data into train/test sets
from sklearn.linear_model import LogisticRegression # logistic model
from sklearn.metrics import accuracy_score # accuarcy
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay

# Settings
pd.set_option('display.max_columns', None) # display all columns
warnings.filterwarnings('ignore') # ignore warnings
%matplotlib inline # To show first the plot
```

```python

connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

## 1. Getting the data

```python
query = '''with cte1 as (select f.film_id, c.name from film f 
join film_category fc
using (film_id)
join category c
using (category_id)), 
cte2 as (select f.film_id, l.name from film f 
join language l
using (language_id))
select f.film_id, cte1.name as category, cte2.name as language, f.release_year, f.rental_rate, f.length, f.rating
from film f
join inventory i
using (film_id)
join rental r
using(inventory_id)
join cte1 cte1 using (film_id)
join cte2 cte2 using (film_id);
'''

data = pd.read_sql_query(query, engine)
data.head()
```

```python
data["film_id"].value_counts()
```

```python
df = data.groupby(data.columns.tolist(),as_index=False).size()
df.head()
```

```python
data.language.value_counts() #we'll drop this
data.release_year.value_counts() #and this, due to single vlaues
```

```python
df = df.drop(columns =['film_id','language', 'release_year'])
df.head()
```

## 2. Cleaning the Data

```python
cat = df.select_dtypes(include=np.object)
num = df.select_dtypes(include=np.number)
```

```python
cat_dummy=pd.get_dummies(cat, columns=["category","rating"])
```

```python
sns.histplot(num['size'], kde=True)
plt.xlabel('Number of Rentals')
plt.ylabel('Frequency')
plt.title('Distribution of Rentals')
plt.show()
```

```python
cat_dummy.sample(100)
```

```python
num.head()
```

```python
mean = num['size'].mean()
median = num['size'].median()
print("Mean is ", mean)
print("Median is ", median)
```

After checking the dates of rental_date column, we can see that the dates go from **2005-05-24 to 2006-02-14**

So all the rentals happened within 8 full months plus 20 days ~8,66 months, so if the value of rentals divided by that is close or higher than 1, the movie has been being rented one or more than once a month.

Given that the median is 16, a lot of films could have been rented more than twice a month, and given that, also once.

```python
#If we are conservatives, we are going to take that that rate has to be equal or higher than one
total_time = 8.66
```

```python
num['probability'] = [1 if x/total_time > 1 else 0 for x in df['size']]
```

## 4. Data Processing

```python
num.head(3)
```

```python
# We don't want to normalize the target
num2 = num.drop(columns=['probability'])

# Selecting the scaler
transformer = StandardScaler().fit(num2)

n = transformer.transform(num2)
n = pd.DataFrame(n) # making it a panda dataframe
n['probability'] = num['probability'] # adding our target
n.columns = num.columns # keeping old columns
n.head()
```

```python
# Merging the two dataframes
X_N = pd.concat([cat_dummy, n], axis=1)
X_N.sample(5)
```

```python
# We need to be careful about when and how we do the split.
Y = X_N['probability']
X = X_N.drop(['probability'], axis=1)
```

## 5. Modeling

```python
# We define train and test for X and Y
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42) 

# test_size = We give 30% for testing and 70% for testing
# random_state = it'll improve the model to divide always the model in the same way
```

```python
# We traing the training dataset using Logistic Regression
logistic = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
# Predictions
predictions = logistic.predict(X_test)
# Converting the numpy array to panda series
pd.Series(predictions).value_counts()
```

```python
# We compare it to the true values from historical data
y_test.value_counts()
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y_test, predictions)

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Greens')
plt.show()
```

## 6. Reporting

```python
# Calculate fpr, tpr, and threshold values for ROC curve
fpr, tpr, thresholds = roc_curve(y_test, predictions, pos_label=1)

# Plot the ROC curve
plt.plot(fpr, tpr, color='red', lw=1.5)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
train_predictions = logistic.predict(X_train)
test_predictions = logistic.predict(X_test)
```

```python
# Add the predicted demand column to the train and test sets
X_train['predicted'] = train_predictions
X_test['predicted'] = test_predictions
```

```python
# Combine the train and test sets
X_pred = pd.concat([X_train, X_test])
X_pred['probability'] = pd.concat([y_train, y_test])
```

```python
# Adding the new column `predicted_demand` to the previous num dataset
num['predicted'] = X_pred['predicted']
```

```python
# Display the entire dataset with the new predicted_demand column
num[['size', 'probability', 'predicted']].sort_values(by='size', ascending=False).head(5)
```
